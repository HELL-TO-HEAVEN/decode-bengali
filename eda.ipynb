```python
import pandas as pd
```

```python
train_df_ = pd.read_csv('bengaliai-cv19/train.csv')
test_df_ = pd.read_csv('bengaliai-cv19/test.csv')
class_map_df = pd.read_csv('bengaliai-cv19/class_map.csv')
sample_sub_df = pd.read_csv('bengaliai-cv19/sample_submission.csv')
```

```python
# Most used vowel diacritic
temp_vowel = train_df_.groupby(by='vowel_diacritic').count()
temp_vowel.reset_index(inplace=True)
vowel_diacritic_map = class_map_df[class_map_df['component_type']=='vowel_diacritic']
vowel_diacritic_count = pd.merge(temp_vowel,vowel_diacritic_map,left_on='vowel_diacritic',right_on='label')
vowel_diacritic_count = vowel_diacritic_count[['vowel_diacritic','image_id','component']]
vowel_diacritic_count.rename(columns={'image_id': "count", 'vowel_diacritic': "vowel_diacritic_label"},inplace=True)
vowel_diacritic_count.sort_values(by=['count'],ascending=False)
```

```python
# Most used root
temp_root = train_df_.groupby(by='grapheme_root').count()
temp_root.reset_index(inplace=True)
root_map = class_map_df[class_map_df['component_type']=='grapheme_root']
root_count = pd.merge(temp_root,root_map,left_on='grapheme_root',right_on='label')
root_count = root_count[['grapheme_root','image_id','component']]
root_count.rename(columns={'image_id': "count", 'grapheme_root': "grapheme_root_label"},inplace=True)
root_count.sort_values(by=['count'],ascending=False)
```

```python
# Most used consonant diacritic
temp_consonant_diacritic = train_df_.groupby(by='consonant_diacritic').count()
temp_consonant_diacritic.reset_index(inplace=True)
consonant_diacritic_map = class_map_df[class_map_df['component_type']=='consonant_diacritic']
consonant_diacritic_count = pd.merge(temp_consonant_diacritic,consonant_diacritic_map,left_on='consonant_diacritic',right_on='label')
consonant_diacritic_count = consonant_diacritic_count[['consonant_diacritic','image_id','component']]
consonant_diacritic_count.rename(columns={'image_id': "count", 'consonant_diacritic': "consonant_diacritic_label"},inplace=True)
consonant_diacritic_count.sort_values(by=['count'],ascending=False)
```
